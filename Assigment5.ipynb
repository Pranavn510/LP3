```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv' , encoding = "unicode_escape")
```

```python
df.head()
```

```python
# Drop rows with missing values (optional)
df_cleaned = df.dropna()
```

```python
# Select numerical columns for clustering
X = df_cleaned[['SALES' , 'QUANTITYORDERED' , 'PRICEEACH']]
```

```python
# Normalize the data (optional, but recommended for K-Means)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Elbow method to find optimal number of clusters
```

```python
from sklearn.cluster import KMeans

K = range(1, 11)  # For example, testing 1 to 10 clusters
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)  # 'data' is your dataset
    inertia.append(kmeans.inertia_)
```

```python
# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-', color='blue')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Choose k based on elbow point (e.g., k=3)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_cleaned['Cluster'] = kmeans.fit_predict(X_scaled)
```
